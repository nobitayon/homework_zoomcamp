```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_0=np.array([5,4])
theta=np.pi/4
mat_rotation=np.array([[np.cos(theta),-np.sin(theta) ],[np.sin(theta),np.cos(theta)] ])
r=0.95
record=[]
record.append(x_0)
for i in range(50):
    x_0=r*np.dot(mat_rotation,x_0)
    record.append(x_0)
record=np.array(record)    
```

```python
fig,ax=plt.subplots(figsize=(10,10))
ax.plot(record[:,0],record[:,1],'bo-')
```

```python
x_0=np.array([10,10])
x_prime=np.array([4,6])
theta=np.pi/4
mat_rotation=np.array([[np.cos(theta),-np.sin(theta) ],[np.sin(theta),np.cos(theta)] ])
record=[]
record.append(x_0)
for i in range(25):
    s_2=r*mat_rotation
    x_0=np.dot(s_2,x_0) - np.dot( (s_2 - np.identity(2) ) ,x_prime)
    record.append(x_0)
record=np.array(record)    
```

```python
fig,ax=plt.subplots(figsize=(10,10))
ax.plot(record[:,0],record[:,1],'bo-')
```
